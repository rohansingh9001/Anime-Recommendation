```python
import pandas as pd
import json
```

```python
data = pd.read_csv("raw_data.csv")
data.head()
```

```python
def clean_excess(string):
    string = string.replace("{{", "{")
    string = string.replace("}}", "}")
    return string

def csv_to_json(data):
    users = []
    for index, row in data.iterrows():
        try:
            anime = row[1]
            anime = anime.replace('""', '"')
            anime = list(anime.split("},{"))

            non_parsable_anime = []

            for i, show in enumerate(anime):
                try:
                    anime[i] = "{" + show + "}"

                    anime[i] = clean_excess(anime[i])

                    anime[i] = json.loads(anime[i])
                except:
                    non_parsable_anime.append(anime[i])

            for error in non_parsable_anime:
                anime.remove(error)

            stats = json.loads(row[2])
            users.append((anime, stats))
        except:
            pass
        
    return users
```

```python
data = csv_to_json(data)
```

```python
USERS = len(data)
len(data)
```

```python
shows = set([])

for animes, stats in data:
    for anime in animes:
        shows.add(anime['anime_title'])

shows = list(shows)
SHOWS = len(shows)
print(SHOWS)
```

```python
table = {}

for show in shows:
    table[show] = [None for user in range(USERS)]
```

```python
for index, data_tuple in enumerate(data):
    animes, stats = data_tuple
    for anime in animes:
        table[anime["anime_title"]][index] = anime["score"]
        
```

```python
table = pd.DataFrame(table)
table.head()
```

```python
table.to_csv('proc_data.csv', index=False)
```
